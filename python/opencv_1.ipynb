```python
fname = './images/b.jpg'

original = cv2.imread(fname, cv2.IMREAD_COLOR)
gray = cv2.imread(fname, cv2.IMREAD_GRAYSCALE)
unchange = cv2.imread(fname, cv2.IMREAD_UNCHANGED)

cv2.imshow('Original', original)
cv2.imshow('Gray', gray)
cv2.imshow('Unchange', unchange)

cv2.waitKey(0)   #  .   
cv2.destroyAllWindows()  #   
```

```python
import cv2

fname_o = './images/b.jpg'
fname_t = './images/b_gray.jpg'
gray = cv2.imread(fname_o, cv2.IMREAD_GRAYSCALE)
cv2.imwrite(fname_t,gray)
img = cv2.imread(fname_t)
cv2.imshow('image',img)
cv2.waitKey(0)   #  
cv2.destroyAllWindows()  #   
```

```python
import cv2
from matplotlib import pyplot as plt # as alias  

img = cv2.imread('./images/a.jpg', cv2.IMREAD_COLOR)

#opencv   r,g,b  
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b,,g, r  Merge

plt.imshow(img2)
plt.xticks([]) # x 
plt.yticks([]) # y 
plt.show()
```
